```
list(movies['genres'].apply(lambda x: x.split("|")))
```

```
genres_list=list(movies['genres'].apply(lambda x: x.split("|")))
genres_list[:3]
```

```
#python flatten list of list
flat_list=[]
for sublist in genres_list:
  for item in sublist:
    flat_list.append(item)
```

```
flat_list
```

```
#중복 데이터 없애기
list(set(flat_list))
```

```
genres_unique=list(set(flat_list))
```

```
len(genres_unique) #총 19개의 장르
```

```
movies['genres'].str.get_dummies(sep='|') #True는 1, False는 0
```

```
genres_dummies=movies['genres'].str.get_dummies(sep='|')
```

```
#데이터 저장하기
genres_dummies.to_pickle('/content/drive/MyDrive/인프런_추천시스템입문/data/ml-latest-small/genres.p')
```

```
#장르 간 상관관계
#두 장르의 관계가 1에 가깝다 : 두 장르가 자주 같이 출현
#두 장르의 관계가 -1에 가깝다 : 두 장르가 아주 드물게 출현, 겹치는 영역이 없음
#두 장르의 관계가 0에 가깝다 : 두 장르 사이의 별 관계가 없다
genres_dummies.corr()
```

```
plt.figure(figsize=(30,15))
sns.heatmap(genres_dummies.corr())
```

```
plt.figure(figsize=(30,15))
sns.heatmap(genres_dummies.corr(), annot=True)
#검은색은 음의 상관관계가 높다.
```
